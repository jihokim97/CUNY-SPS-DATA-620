Dataset:
The file lesmis.gml contains the weighted network of coappearances of characters in Victor Hugo's novel "Les Miserables". Nodes represent characters as indicated by the labels and edges connect any pair of characters that appear in the same chapter of the book. The values on the edges are the number of such coappearances.

- D. E. Knuth, The Stanford GraphBase: A Platform for Combinatorial Computing, Addison-Wesley, Reading, MA (1993). Retreived from http://www-personal.umich.edu/~mejn/netdata/. 



```python
# Libraries
import networkx as nx
import pylab as plt
```

```python
G = nx.read_gml('lesmis.gml')
G
```

### Info about Graph

```python
nx.info(G)
```

```python
G.nodes
```

### Analysis
- Diameter
- Density
- most influential


```python
# Diameter
print(nx.is_connected(G))

diameter = nx.diameter(G)
print('Network diameter:', diameter)
```

```python
# Density
density = nx.density(G)
print('Network density:', density)
```

```python
# Most Influential
most_influential = nx.degree_centrality(G)
for i in sorted(most_influential, key=most_influential.get, reverse=True):
    print(i, most_influential[i])
```

The diameter was 5 which indicates the furthest two nodes were 5 and the density was found to be 0.0686 which indicates that the network is not dense. All the graphs were connected so I did not have to create a subgroup to find a diameter. The most influential person from the novel was Valjean which is not surprising since he is the main character of the novel.
